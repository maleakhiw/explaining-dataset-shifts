## Barplot explanations

The barplots depict the degree how each concept is affected by various type of shifts. Higher value indicates that the concept is more shifted relative to others.

```python
list_shifts = ["Knockout", "Concept (Floor)", "Concept (Wall)",
              "Concept (Object + Scale)", "Rotation", "Shear", "Flip", 
               "All", "Gaussian", "FGSM"]
list_labels = ["CBSDs", "CBSDh"]
list_shift_str = ["ko", "floor", "wall", "object_scale", "rotation", 
                  "shear", "flip", "all", "gaussian", "adversarial"]
list_dict_results = []
for shift_str in list_shift_str:
    temp = []
    for label in list_labels:
        temp.append(load_result(shift_str, label, False, "3dshapes"))
    list_dict_results.append(temp)

list_shift_intensity = [ShiftIntensity.Medium, ShiftIntensity.Large]
list_shift_prop = [1.0]
list_test_sample = [1000, 10000]
concept_names = ["floor", "wall", "object", "scale", "shape", "orientation"]
```

```python
for shift, dict_result in zip(list_shifts, list_dict_results):
    display(Markdown(f"# {shift}"))
    
    dict_cbsds = dict_result[0]
    dict_cbsdh = dict_result[1]
    
    barplot_test_statistics(dict_cbsds, dict_cbsdh,
                           concept_names, list_shift_intensity,
                           list_shift_prop, list_test_sample)
```
